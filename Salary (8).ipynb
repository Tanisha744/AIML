```python
# to fix BasePay object error
dataframe['BasePay'] = pd.to_numeric(dataframe['BasePay'],errors="coerce")
```

# FIND MINIMUM,MAXIMUM AND AVERGE BASEPAY

```python
print("min basepay :",dataframe["BasePay"].min())
print("max basepay :",dataframe["BasePay"].max())
print("avg basepay :",dataframe["BasePay"].mean())
```

# REPLACE 'NOT PROVIDED' IN  EMPLOYEENAME COLUMN TO NaN

```python
dataframe.replace("Not Provided","NaN")
```
