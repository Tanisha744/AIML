```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("adult.csv")
df.replace("?",pd.NA,inplace=True)
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.countplot(data=df)
plt.xticks(rotation=90)
plt.show()
```
