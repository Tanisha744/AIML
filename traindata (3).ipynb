# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
import matplotlib.pyplot as plt

# Passenger data
sex = ["male","female"]
survived = data["Survived"].value_counts()
# Create a bar chart
plt.figure(figsize=(5,5))
plt.bar(sex,survived)
plt.xlabel("Sex")
plt.ylabel("Survived")
plt.title("Better Chance")
# plt.grid(axis="y", linestyle="--", alpha=0.7)

# Display the chart
plt.show()
```

# WHICH PASSENGER CLASS HAS BETTER CHANCE OF SURVIVAL(FIRST,SECOND,OR THIRD CLASS)?

```python
import matplotlib.pyplot as plt

# Passenger data
classes = ["First Class", "Second Class", "Third Class"]
survived = data["Survived"].value_counts()
# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(classes,survived,color="red","blue")
plt.xlabel("Passenger Class")
plt.ylabel("Survived")
plt.title("Passenger Distribution by Class")
# Display the chart
plt.show()
```
